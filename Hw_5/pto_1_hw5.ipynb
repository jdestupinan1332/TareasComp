```python
%pylab inline
from scipy.fftpack import fft, fftfreq
import numpy as np
```

```python
data=loadtxt("sampled+ma0844az_1-1+_data.txt") 
times = arange(0,400,1.0) #x axis in the plot
column = data[:,0] #y axis in the plot
dimdata = shape(data) #dimensions of column
```

```python
plot (times, column)
```

```python
fft_x = fft(column) / dimdata[0] # FFT Normalized
freq = fftfreq(dimdata[0], 1) # Recuperamos las frecuencias
```

```python
fft_x_shifted = np.fft.fftshift(fft_x)
freq_shifted = np.fft.fftshift(freq)
plot (freq_shifted,np.abs(fft_x_shifted))
```

```python
espectropot =  (fft_x_shifted.real**2 + fft_x_shifted.imag**2) #espectro de potencias
```

```python
maxespect = (-espectropot).argsort()[:400]
```

```python
espectropot[155]
```

```python
cleaned = zeros(400)
for item in maxespect:
    cleaned[item] = fft_x[item]
    
        
```

```python
plot (np.abs(cleaned))
```

```python
inverse = ifft(cleaned)*400
```

```python
plot (inverse)
```

```python
chi2 = 0.0
reconstruct = inverse.real
for i in range(len(inverse)):
    chi2 = chi2 + (reconstruct[i] - column[i])**2
    
    
chi2/400
```
