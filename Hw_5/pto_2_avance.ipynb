```python
from scipy.interpolate import interp1d
import numpy as np
%pylab inline
data=loadtxt("sparse_sample_monthrg.dat")
anno=data[:,0] #list of years
mes=data[:,1] #list of months
dias=data[:,2] #list os days, NOT USED
manchas = data[:,3] #list of sunspots

data_n=[]
tiempo=[]
#creates the list useful measured data (!=-99, with the correct time)
for i in range(len(manchas)):
    if(manchas[i] >= 0):
        data_n.append(manchas[i])
        tiempo.append(anno[i]+(mes[i]-1)/12)
t = arange(1613,1995,0.5) #array of real time
```

```python
#linear interpolation
f_l=interp1d(tiempo,data_n,kind='linear')
plot(tiempo, data_n,'ko')
plot(t,f_l(t))
```

```python
#cubic interpolation
f_cu=interp1d(tiempo,data_n,kind='cubic')
plot(tiempo, data_n,'ko')
plot(t,f_cu(t))
```

```python
#constant interpolation
f_const=interp1d(tiempo,data_n,kind='zero')
plot(tiempo,data_n,'ko')
plot(t,f_const(t))
```

```python
from scipy.fftpack import fft, fftfreq

#Frequencies
freq=fftfreq(shape(t)[0])
```

```python
#fourier Transform
fft_manchas_lin = fft(f_l(t))/shape(t)[0]
fft_manchas_cub = fft(f_cu(t))/shape(t)[0]
fft_manchas_const = fft(f_const(t))/shape(t)[0]
```

```python
#Power Spectra
powersp_lin=fft_manchas_lin.real**2+fft_manchas_lin.imag**2
powersp_cub=fft_manchas_cub.real**2+fft_manchas_cub.imag**2
powersp_const=fft_manchas_const.real**2+fft_manchas_const.imag**2
```

```python
# plots power spectra
plot (np.fft.fftshift(freq),log(np.fft.fftshift(powersp_lin)), color='red')
plot (np.fft.fftshift(freq),log(np.fft.fftshift(powersp_cub)),color='green')
plot (np.fft.fftshift(freq),log(np.fft.fftshift(powersp_const)),color='blue')
```

```python
#range of frequencies
pdo_2=1/2.0
pdo_20=1/20.0
```

```python
#Reconstruction of linear data
new_fft_lin=[]
for i in range(len(t)):
    if(freq[i]<-pdo_2) or (freq[i]>pdo_2):
        fft_manchas_lin[i]=0
    if(freq[i]>-pdo_20) and (freq[i]< pdo_20):
        fft_manchas_lin[i]=0
    
    new_fft_lin.append(fft_manchas_lin[i])
new_data_lin=ifft(new_fft_lin)*shape(t)[0]
```

```python
plot(t, new_data_lin)
plot(t,f_l(t))
legend(['reconstruct', 'interpolation'])
```

```python
#Reconstruction of cubic data
new_fft_cub=[]
for i in range(len(t)):
    if(freq[i]<-pdo_2) or (freq[i]>pdo_2):
        fft_manchas_cub[i]=0
    if(freq[i]>-pdo_20) and (freq[i]< pdo_20):
        fft_manchas_cub[i]=0
        
    new_fft_cub.append(fft_manchas_cub[i])
new_data_cub=ifft(new_fft_cub)*shape(t)[0]

```

```python
plot(t, new_data_cub)
plot(t,f_cu(t))
legend(['reconstruct', 'interpolation'])
```

```python
#reconstruction of constant data
new_fft_co=[]
for i in range(len(t)):
    if(freq[i]<-pdo_2) or (freq[i]>pdo_2):
        fft_manchas_const[i]=0
    if(freq[i]>-pdo_20) and (freq[i]< pdo_20):
        fft_manchas_const[i]=0
        
    new_fft_co.append(fft_manchas_const[i])
new_data_co=ifft(new_fft_co)*shape(t)[0]
```

```python
plot(t, new_data_co)
plot(t,f_const(t))
legend(['reconstruct', 'interpolation'])
```

```python
#estimación del periodo lineal
lin = fft(new_data_lin)
imaxl = (-lin).argsort()[:2]
1/freq[imaxl][1]
```

```python
#estimación del periodo cúbica
cub = fft(new_data_cub)
imaxc = (-cub).argsort()[:2]
1/freq[imaxc][1]
```

```python
#estimación del periodo constante
co = fft(new_data_co)
imaxco = (-co).argsort()[:2]
1/freq[imaxco][1]
```
