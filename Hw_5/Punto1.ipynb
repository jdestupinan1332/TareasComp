```python
%pylab inline
from scipy.fftpack import fft, fftfreq
import numpy as np

data=loadtxt("sampled+ma0844az_1-1+_data.txt") 
dimdata = shape(data) #dimensions of column
```

```python
def givecolumn(data, i):
    column = data[:,i]
    return column

def fourier(column, data):
    fft_x = fft(column) / dimdata[0] # FFT Normalized
    freq = fftfreq(dimdata[0], 1) # Recuperamos las frecuencias 
    return fft_x, freq

def powerspectrum(fft_x):
    powersp = fft_x.real**2 + fft_x.imag**2
    return powersp

def searchmax(powersp, fft_x):
    maxespect = (-powersp).argsort()[:10]
    cleaned = zeros(400, dtype='complex')
    for item in maxespect:
        cleaned[item] = fft_x[item]
    return cleaned

def reconstruct(cleaned):
    recons = ifft(cleaned)*dimdata[0]
    return recons

def chisq(recons, column):
    chi2 = 0.0
    for i in range(dimdata[0]):
        chi2 = chi2 + (recons[i] - column[i]).real**2
    print chi2/dimdata[0]
```

```python
for i in range(dimdata[1]):
    column = givecolumn(data,i)
    fft_x, freq = fourier(column,data)
    powersp = powerspectrum(fft_x)
    cleaned = searchmax(powersp, fft_x)
    recons = reconstruct(cleaned)
    chisq(recons, column)
    plot (recons)
title('reconstruction')
show()
```
